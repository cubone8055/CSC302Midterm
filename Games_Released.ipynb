```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("/content/drive/Shareddrives/302 Project/Data/GameSales.csv")
df = df.groupby('console')['title'].count()
df = df.sort_values(ascending=False)
df = df.reset_index()
df = df[1:13]

ax =  plt.bar(df['console'],df['title'])
ax = plt.gca()
fig = plt.gcf()
print(ax.get_yticks())
ax.set_ylabel("Amount of Titles")
ax.set_xlabel("Consoles")


```
