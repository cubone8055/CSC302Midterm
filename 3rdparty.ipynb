```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("/content/drive/Shareddrives/302 Project/Data/GameSales.csv")
df2 = df.loc[df['title'] == 'Minecraft']
df2 = df2.loc[(df['console'] == 'XOne') | (df['console'] == 'PS3') | (df['console'] == 'PS4') | (df['console'] == 'WiiU')]

ax =  plt.bar(df2['console'],df2['total_sales'])
ax = plt.gca()
fig = plt.gcf()
print(ax.get_yticks())
ax.set_ylabel("Minecraft Sales (Millions)")
ax.set_xlabel("Consoles")

plot = df2.plot.pie(y='total_sales', figsize=(7, 7), legend=True, labels=df2['total_sales'])
plot.legend(df2['console'])
plot.set_xlabel("Minecraft Sales (Millions)")
plot.set_ylabel("")
```
